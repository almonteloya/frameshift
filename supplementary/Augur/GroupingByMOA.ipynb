```python
import pandas as pd
from glob import glob
```

```python
def parse_out(path):
	files = glob(path + "/*.csv")
	dfs = {}
	for f in files:
		base_name = f.split("/")[-1].split(".csv")[0]
		plate, drug = base_name.split("__")
		# drug, plate 
		df = pd.read_csv(f)
		df['plate'] = plate
		df.set_index(['plate', 'drug'], inplace=True)
		dfs[(plate, drug)] = df.iloc[0,:]
	dfs = pd.concat(dfs, axis=1).T.reset_index().rename(columns={"level_0": "plate", "level_1": "drug"})
	return(dfs)


path = "/home/ubuntu/frameshift-1/data/augur_out/scores"
scores = parse_out(path)

scores = scores.groupby('drug').mean(numeric_only=True)
print(scores.shape)
scores.head()
```
